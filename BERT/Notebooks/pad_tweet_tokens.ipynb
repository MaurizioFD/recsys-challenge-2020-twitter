### Open files to be read and perform the padding (and save the results)

```python
tokens_file = open(TOKENS_FILE, "r")
```

```python
def save_tweet(index, text):
    line = index + ','
    for item in text:
        line += item 
```

```python
%%time

# ABOUT ~1h 50min MINUTES EXECUTION ON THE WHOLE DATASET

# COMMENTED CODE IS FOR DEBUGGING (TO CHECK THAT ALL THE OUTPUTS HAVE THE SAME LENGTH)

# ignore header
line = tokens_file.readline()

finished = False
i = 0

while not finished:  # and i < 10:
    
    if i % 1000000 == 0:
        print(i)
    
    line = str(tokens_file.readline())
    
    if line != '':
        
        #print(line)
        
        line = line.replace("\\n",'').replace('\n','').replace("\\t",'\t')
        splitted_line = line.split(',')
        splitted_tokens = splitted_line[1].split('\t')  # take only the tokens (splitted_line[0] is the tweet id)
        
        #splitted_line = []
        for j in range(len(splitted_tokens), max_tweet_length):
            line += PAD
            #splitted_line.append(PAD)
        
        line += '\n'
        
        #if (len(splitted_line)+len(splitted_tokens) != max_tweet_length):
        #    print('Error')
        #    break
        
        result_file.write(line)
        
    else:
        finished = True

    i += 1
```
