```python
import pandas as pd
import numpy as np
```

```python
from transformers import BertTokenizer

tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-cased')
```

```python
tweet_df = pd.read_csv(PATH)
tweet_df.head()
```

```python
tweet_dict = {}
```

```python
# Declaring two lambdas in order to cast a string to a numpy array of integers
f_to_int = lambda x: int(x)
f_int = lambda x: list(map(f_to_int, x.split('\t')))
```

```python
for i in range(len(tweet_df)):
  tweet_id = tweet_df.iloc[i][0]
  tweet_raw_text =  tweet_df.iloc[i][1]
  tweet_dict[tweet_id] = f_int(tweet_raw_text)
```

```python
for k in tweet_dict.keys():
  print(k)
  print(tokenizer.decode(tweet_dict[k]))
```
