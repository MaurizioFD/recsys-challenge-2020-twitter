# Drop tokens from the longest tweet if it exceeds max BERT length (that is 512)

```python
import time
```

### Constants

```python
LONGEST_TWEET_ID = 63888121
```

### Paths

```python
TWEET_ID = "tweet_features_tweet_id"
TWEET_TOKENS = "tweet_features_text_tokens"

TWEET_TOKENS_FILE = "tweet_tokens/text_tokens_all_no_escaped_chars.csv"

RESULT_PATH = "tweet_tokens/text_tokens_all_reduced_longest_tweet.csv"
```

```python
result_file = open(RESULT_PATH, "w+")
```

### Write the header (columns names)

```python
result_file.write(TWEET_ID + ',' + TWEET_TOKENS + "\n")
```

### Open files to be read and perform the substitution

```python
tokens_file = open(TWEET_TOKENS_FILE, "r")
```

```python
def save_tweet(index, text):
    string = index + ','
    for item in text:
        string += '\t' + item 
    result_file.write(string)
```

```python
%%time

# ~4 MINUTES EXECUTION ON THE WHOLE DATASET

# ignore header
tokens_file.readline()

finished = False
i = 0

while not finished:  # and i < 1:
    
    if i % 1000000 == 0:
        print(i)
            
    line = str(tokens_file.readline())
    
    if line != '':
        
        if i == LONGEST_TWEET_ID:
            line = line.split('\t')  # replace("\\n",'').replace("\\t",'\t')
            print(line)

            for j in range(18):
                line.pop(1)  # avoid popping the element in pos 0 (containing tweet_id and CLS token)

            print(line)
            save_tweet(str(i), line)
        
        else:
            result_file.write(line)
        
    else:
        finished = True

    i += 1
```

```python
tokens_file.close()

result_file.close()
```
